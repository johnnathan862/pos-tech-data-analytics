# Set-up

```python
from google.cloud import bigquery
from google.oauth2 import service_account

import pandas as pd
import pandas_gbq
```

```python
# Objetivo com as credenciais
credentials = service_account.Credentials.from_service_account_file(
    filename='credentials_gcp_postech.json',
    scopes=['https://www.googleapis.com/auth/bigquery']
)
```

# Leitura de dados

```python
query = '''
SELECT * FROM `pos-tech-project.pnad.pnad-covid-112020` LIMIT 10
'''
```

```python
df =  pandas_gbq.read_gbq(credentials=credentials, query_or_table=query)
```
