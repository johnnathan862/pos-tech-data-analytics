```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import missingno as mo
```

```python
df = pd.read_csv(
    "A150850189_28_143_208 (1).csv", 
    encoding='latin', 
    sep=';', 
    skiprows=3, 
    skipfooter=12, 
    decimal=',', 
    na_values=['-']
)
```

```python
df.dtypes
```

```python
df
```

```python
mo.matrix(df=df)
```

```python
axis = df.sort_values(
    by='2021/Set', ascending=True, na_position='last'
).plot(
    x='Unidade da Federação', y='2021/Set', figsize=(10,5), kind='barh', fontsize=8
)

axis.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
axis.set_xlabel("Qtd.")
axis.set_title("Gráfico com barras horizantais")
plt.legend(['2021/Set'], loc='lower right')

plt.show()
```

```python
# fig = plt.figure(figsize=(10,5))
# axis1 = fig.add_axes([0,0,1,1])
# axis1.bar(x=df['Unidade da Federação'], y=df['2021/Set'])
# 

fix, axis = plt.subplots(figsize=(10,5))

# Df ordenado
df_ordered = df.sort_values(
    by='2021/Set', ascending=True, na_position='last'
)

# Eixo do gráfico de barras horizontais
axis.barh(
    df_ordered['Unidade da Federação'], 
    df_ordered['2021/Set']
)

axis.set_ylabel("Unidade da Federação", fontsize=10)
axis.set_xlabel("Qtd.", fontsize=10)
axis.set_title("Gráfico de qtd. por UF", fontsize=12)
axis.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
axis.grid(True)
axis.legend(['2021/Set'])

# Eixo do gráfico de pizza
axis2 = fix.add_axes([0.57,0.25,0.4,0.5])
# axis2.pie([10,25,65], labels=['azul','verde','amarelo'])
axis2.pie(df_ordered.fillna(0)['2021/Set'], labels=df_ordered['Unidade da Federação'], autopct='%1.1f%%')

plt.show()
```
